# soru 3
authon field george orwell olan document genre field Sciene_Fiction olarak güncelle ve yazdır

```python
tablo.update_many(
    {'author': 'George Orwell'},
    {'$set': {'genre': 'Science Fiction'}} 
    )
```

# soru 4
kitabın basım yılının 1950 den fazla olanları yazdırın. for döngüsü ile anlaşılabilir bir format oluşturun. bunun için '$gte' methodunu kullanın

```python
basim_yil = tablo.find({'year': {'$gt': 1950}}) 
print("basım yılı 1950'den sonra olan kitaplar:")
for veri in basim_yil:
    print(f"""***
        kitap türü: {veri['genre']}

        kitap adı: {veri['title']}

        kitap yazarı: {veri['author']}
        kitap basım yılı: {veri['year']}
------------------------------------------
    """)
```
